**Data Analysis and Visualization**

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel',hue = 'is_canceled', data = df, palette = 'Blues')
plt.title('Reservation status in different hotels',size= 18)
plt.show()
```

```python
x = df[df['hotel'] == 'Resort Hotel']
x['is_canceled'].value_counts(normalize = True)
```

```python
y = df[df['hotel'] == 'City Hotel']
y['is_canceled'].value_counts(normalize = True)
```

```python
x = x.groupby('reservation_status_date')[['adr']].mean()
y = y.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in city and Resort Hotel',fontsize = 28)
plt.plot(x.index,x['adr'],label = 'resort Hotel')
plt.plot(y.index,y['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=28)
# Calculate the data for plotting
monthly_adr = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
# Use keyword arguments for barplot
sns.barplot(x='month', y='adr', data=monthly_adr)
plt.legend(fontsize=20)  # This may not be necessary for a bar plot.
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize= (8,8))
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```
